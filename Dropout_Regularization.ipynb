```python
from sklearn.preprocessing import LabelEncoder
l1 = LabelEncoder()
df[60] = l1.fit_transform(df[60])
```

```python
x = df.drop(60,axis = 1)
y = df[60]
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25)
x_train
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(60,input_dim = 60,activation = 'relu'),
    keras.layers.Dense(30,activation = 'relu'),
    keras.layers.Dense(15,activation = 'relu'),
    keras.layers.Dense(1,activation = 'sigmoid')
])
model.summary()
```

```python
model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)
model.fit(x_train,y_train,epochs =10,batch_size = 32)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
y_pred = model.predict(x_test).reshape(-1)

ans = []
for actual in y_pred:
  if actual>=0.5:
    ans.append(1)
  else:
    ans.append(0)

print(ans)

```

```python
model_drop = keras.Sequential([
    keras.layers.Dense(60,input_dim = 60,activation = 'relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30,activation = 'relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15,activation = 'relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1,activation = 'sigmoid')
])
model_drop.summary()
```

```python
model_drop.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)
model_drop.fit(x_train,y_train,epochs =10,batch_size = 32)
```

```python
y_pred = model_drop.predict(x_test).reshape(-1)

ans = []
for actual in y_pred:
  if actual>=0.5:
    ans.append(1)
  else:
    ans.append(0)

print(ans)

```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,ans))
```
